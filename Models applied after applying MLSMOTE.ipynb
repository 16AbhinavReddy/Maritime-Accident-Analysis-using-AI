```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import LinearSVC
from sklearn.tree import DecisionTreeClassifier
from catboost import CatBoostClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
from sklearn.metrics import hamming_loss
from MLSMOTE import get_minority_instace, MLSMOTE
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv("data_preprocess.csv")
```

```python
x = data["Overall Info"]
y = data[data.columns[6:17]]
```

```python
y
```

```python
vectorizer = TfidfVectorizer(max_features=700, max_df=0.85)
vectorizer.fit(x)
X_transformed = vectorizer.transform(x)
X_transformed_df = pd.DataFrame(X_transformed.toarray())
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_transformed_df, y, test_size=0.15, random_state=101)
```

```python
x_train
```

```python
X_sub, y_sub = get_minority_instace(x_train, y_train)
X_res, y_res = MLSMOTE(X_sub, y_sub, 500)
```

```python
print("Original X shape:", x_train.shape)
print("Resampled X shape:", X_res.shape)
print("Original y shape:", y_train.shape)
print("Resampled y shape:", y_res.shape)
```

```python
model = OneVsRestClassifier(CatBoostClassifier(iterations=100, depth=6, learning_rate=0.1, verbose=False))
```

```python
model.fit(X_res, y_res)
```

```python
y_pred = model.predict(x_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
loss = hamming_loss(y_test, y_pred)
```

```python
loss
```

```python
model1 = OneVsRestClassifier(LinearSVC())
```

```python
model1.fit(X_res, y_res)
```

```python
y_pred1 = model1.predict(x_test)
```

```python
print(classification_report(y_test, y_pred1))
```

```python
loss1 = hamming_loss(y_test, y_pred1)
```

```python
loss1
```

```python
model2 = OneVsRestClassifier(LogisticRegression())
```

```python
model2.fit(X_res, y_res)
```

```python
y_pred2 = model2.predict(x_test)
```

```python
print(classification_report(y_test, y_pred1))
```

```python
loss2 = hamming_loss(y_test, y_pred2)
```

```python
loss2
```

```python
model3.fit(X_res, y_res)
```

```python
y_pred3 = model3.predict(x_test)
```
